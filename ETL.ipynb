```python
#Importo la libreria de pandas para poder leer los csv
import pandas as pd
```

```python
#Primero trabajo con los csv de las plataformas respectivamente.
amazon_df = pd.read_csv("datasets/amazon_prime_titles.csv")
disney_df = pd.read_csv("datasets/disney_plus_titles.csv")
hulu_df = pd.read_csv("datasets/hulu_titles.csv")
netflix_df = pd.read_csv("datasets/netflix_titles.csv")
amazon_df.head(3)
```

```python
#Una de las consignas es generar un nuevo 'id', añadir que luego vendra bien para diferenciar cada  fila y unificarlos.
amazon_df.insert(loc=0, column="id", value="a" + amazon_df["show_id"])
disney_df.insert(loc=0, column="id", value="d" + disney_df["show_id"])
hulu_df.insert(loc=0, column="id", value="h" + hulu_df["show_id"])
netflix_df.insert(loc=0, column="id", value="n" + netflix_df["show_id"])
disney_df.tail(5)
```

```python
# Unificamos los 4 datasets para proceder con el resto de transformaciones que igualmente se estarían aplicando en los 4 por separado.
movies_df = pd.concat([amazon_df,disney_df,hulu_df,netflix_df], axis=0)
movies_df.reset_index(inplace=True, drop=True) # Restablezco el index por si acaso.
movies_df.tail(3)
```

```python
# Reviso si hay duplicados.
movies_df.duplicated().sum()
```

```python
# Siguiendo las consignas se estaria cambiando los valores NaN en rating(concerniente a edad permitida para ver las peliculas) por "G"( Para todas las edades).
#movies_df.isnull().sum() # ratings 864 null previo
movies_df["rating"].fillna("G", inplace=True)
movies_df.isnull().sum()
```

```python
# Se estaría cambiando el formato de fecha por AAAA-mm-dd
movies_df["date_added"] = pd.to_datetime(movies_df["date_added"])
movies_df.head(3)
```

```python
#Cambiamos todos los datos a minúsculas , incluidos rating(ya modificado con "G") y duration(el cual mas tarde se modificara igualmente).
movies_df["type"] = movies_df["type"].str.lower()
movies_df["title"] = movies_df["title"].str.lower()
movies_df["director"] = movies_df["director"].str.lower()
movies_df["cast"] = movies_df["cast"].str.lower()
movies_df["country"] = movies_df["country"].str.lower()
movies_df["rating"] = movies_df["rating"].str.lower()
movies_df["duration"] = movies_df["duration"].str.lower()
movies_df["listed_in"] = movies_df["listed_in"].str.lower()
movies_df["description"] = movies_df["description"].str.lower()
movies_df
```

```python
#Consigna concerniente a separar el valor duration en dos columnas (duration_int y duration_type).
columns = movies_df["duration"].str.split(" ", n=1, expand=True)
columns
```

```python
#Aquí genero las columnas en el data set añadiendo los valores separados como también el formato. Añadir que se encontraron 4 valores únicos en duration_type('min','seasons'.'season','NaN) por ello decidí convertir los 'season' en 'seasons'.
movies_df["duration_int"] = columns[0]
movies_df["duration_type"] = columns[1]
movies_df["duration_int"] = movies_df["duration_int"].astype("Int64")
movies_df["duration_type"] = movies_df["duration_type"].str.replace("season", "seasons")
movies_df["duration_type"] = movies_df["duration_type"].str.replace("seasonss", "seasons")
movies_df.head(1)
```

```python
# Me deshago de las columnas que ya no considero útiles.
movies_df= movies_df.drop(columns=['show_id','duration'])
movies_df
```

```python
#Procedo a importar los csv de calificaciones por parte de los usuarios para trabajar con ellos.
df1=pd.read_csv("datasets/1.csv")
df2=pd.read_csv("datasets/2.csv")
df3=pd.read_csv("datasets/3.csv")
df4=pd.read_csv("datasets/4.csv")
df5=pd.read_csv("datasets/5.csv")
df6=pd.read_csv("datasets/6.csv")
df7=pd.read_csv("datasets/7.csv")
df8=pd.read_csv("datasets/8.csv")
df8.head()
```

```python
#Unifico los datasets para generalizar las modificaciones e igualmente terminar por extraer los score (aquí llamado 'rating').
rating_users= pd.concat([df1,df2,df3,df4,df5,df6,df7,df8],axis=0)
rating_users
```

```python
#Cambio el dato timestamp, primero con lectura en segundos y luego a la fecha real que ocupan.
rating_users['timestamp']=pd.to_datetime(rating_users['timestamp'],unit='s')
rating_users['timestamp'] = rating_users['timestamp'].dt.date
rating_users  
```

```python
#Cambio el 'movieid' y 'rating' a 'id' y score' respectivamente para poder relacionarlo con la tabla de movies_df
rating_users.rename(columns={'movieId':'id','rating':'score'}, inplace= True)
rating_users
```

```python
# Genero el promedio de score x película con el id correspondiente de la película.
aux= rating_users.groupby('id')['score'].mean()
aux
```

```python
#Unifico la serie 'aux' con el dataset movies_df. Añadiéndole finalmente la columna de 'score' y creando el dataset moviescores.
moviescores=movies_df.merge(aux.rename('score'), left_on='id', right_index=True)
moviescores.tail(3)
```

```python
#Aquí veo los valores nulos que luego tratare en el EDA
moviescores.isnull().sum()
```

```python
#Retiro este dataset en limpio para luego analizarlo sin modificaciones en el EDA
moviescores.to_csv('datasets/platformdata.csv',index=False,sep=';')
```
