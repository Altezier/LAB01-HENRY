```python
import pandas as pd
```

```python
amazon_df = pd.read_csv("datasets/amazon_prime_titles.csv")
disney_df = pd.read_csv("datasets/disney_plus_titles.csv")
hulu_df = pd.read_csv("datasets/hulu_titles.csv")
netflix_df = pd.read_csv("datasets/netflix_titles.csv")
amazon_df.head()
```

```python
amazon_df.insert(loc=0, column="id", value="a" + amazon_df["show_id"])
disney_df.insert(loc=0, column="id", value="d" + disney_df["show_id"])
hulu_df.insert(loc=0, column="id", value="h" + hulu_df["show_id"])
netflix_df.insert(loc=0, column="id", value="n" + netflix_df["show_id"])
disney_df.tail(5)
```

```python
movies_df = pd.concat([amazon_df,disney_df,hulu_df,netflix_df], axis=0)
movies_df
```

```python
movies_df.isnull().sum() # ratings 864 null
movies_df["rating"].fillna("G", inplace=True)
movies_df.isnull().sum()
```

```python
movies_df["date_added"] = pd.to_datetime(movies_df["date_added"])
movies_df.head(3)
```

```python
movies_df["type"] = movies_df["type"].str.lower()
movies_df["title"] = movies_df["title"].str.lower()
movies_df["director"] = movies_df["director"].str.lower()
movies_df["cast"] = movies_df["cast"].str.lower()
movies_df["country"] = movies_df["country"].str.lower()
movies_df["rating"] = movies_df["rating"].str.lower()
movies_df["duration"] = movies_df["duration"].str.lower()
movies_df["listed_in"] = movies_df["listed_in"].str.lower()
movies_df["description"] = movies_df["description"].str.lower()
movies_df
```

```python
movies_df['duration'].isnull().sum()
```

```python
columns = movies_df["duration"].str.split(" ", n=1, expand=True)
columns
```

```python
movies_df["duration_int"] = columns[0]
movies_df["duration_type"] = columns[1]
movies_df["duration_int"] = movies_df["duration_int"].astype("Int64")
movies_df.tail(5)
```

```python
median=movies_df['duration_int'].median(skipna=True) #482 null on duration_int
median
```

```python
movies_df['duration_int']=movies_df['duration_int'].fillna(85)
```

```python
movies_df= movies_df.drop(columns=['show_id','duration'])
movies_df
```

```python
df1=pd.read_csv("datasets/1.csv")
df2=pd.read_csv("datasets/2.csv")
df3=pd.read_csv("datasets/3.csv")
df4=pd.read_csv("datasets/4.csv")
df5=pd.read_csv("datasets/5.csv")
df6=pd.read_csv("datasets/6.csv")
df7=pd.read_csv("datasets/7.csv")
df8=pd.read_csv("datasets/8.csv")
df8.head()
```

```python
rating_users= pd.concat([df1,df2,df3,df4,df5,df6,df7,df8],axis=0)
rating_users
```

```python
rating_users['timestamp']=pd.to_datetime(rating_users['timestamp'],unit='s')
rating_users['timestamp'] = rating_users['timestamp'].dt.date
rating_users  
```

```python
rating_users.to_csv('datasets/rating_users.csv',index=False,sep=';')
```

```python
rating_users.rename(columns={'movieId':'id','rating':'score'}, inplace= True)
rating_users
```

```python
aux= rating_users.groupby('id')['score'].mean()
aux= aux.to_frame()
aux['id']=aux.index
aux=aux.reset_index(drop=True)
aux.dtypes
```

```python
moviescores=pd.merge(movies_df,aux)
moviescores
```

```python
moviescores.to_csv('datasets/moviescores.csv',index=False,sep=';')
```
